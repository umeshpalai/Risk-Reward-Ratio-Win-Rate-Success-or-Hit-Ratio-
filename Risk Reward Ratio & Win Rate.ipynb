### Risk Reward Ratio (RRR) 

Expected Gain of an Investment against the Risk of Loss

RRR = Maximum profit $/$ Maximum risk of loss

RRR = Take Profit Distance $/$ Stop Loss Distance

If your enetry point is at 100. Your Take profit at 180 & Stop loss at 60. Your RRR will be 2:1  
(Take Profit Distance is 180-100=80 $/$ Stop Loss Distance 100-60=40)


### Win Rate (Sucess/Hit ratio)

Your win rate is how many trades you win out of all your total trades. 

Win Rate (Sucess/Hit Ratio)= No of profitable tardes / Total no of trades

Let's say your total no of trade is 100, out of that 50 trades are profitable , so your win rate is 50%

##### Why RRR is so important?

<b> The RRR really tells you how big your win rate has to be in ordr to make money.

    - If your RRR is 1:1, then win rate has to be more than 50% to make money in long run. 
      (For every losing trade, you need to have one wining trade for break even whcih is 50% wining rate)
